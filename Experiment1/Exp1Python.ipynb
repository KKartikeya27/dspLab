```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
import cmath as cm
import scipy.signal
#from scipy.signal import freqz
```

Question 1

```python
x_val = np.arange(-3, 7)

unit_impulse = [1 if x == 0 else 0 for x in x_val]
unit_step = [1 if x >= 0 else 0 for x in x_val]
unit_ramp = [x if x >= 0 else 0 for x in x_val]

plt.stem(x_val, unit_impulse)
plt.title('Unit Impulse Signal')
plt.xlabel('n ---> ')
plt.ylabel('Î´[n] --->')
plt.show()

plt.stem(x_val, unit_step)
plt.title('Unit Step Signal')
plt.xlabel('n ---> ')
plt.ylabel('u[n] --->')
plt.show()

plt.stem(x_val, unit_ramp)
plt.title('Unit Ramp Signal')
plt.xlabel('n ---> ')
plt.ylabel('r[n] --->')
plt.show()
```

```python
x = np.arange(-50, 50)

y1 = [20 * 0.9 ** n for n in x]
y2 = [0.2 * 1.2 ** n for n in x]
# x = np.arange(-5, 5, 0.1)
y3 = [pow(-0.8, pow(n, 4)) for n in x]
y4 = [-4 * 0.8 ** n for n in x]


plt.figure(figsize=(10, 10))
plt.subplot(411)
plt.stem(x, y1)
plt.xlabel('n ---> ')
plt.ylabel('20[0.9]^n')
# plt.show()

plt.subplot(412)
plt.stem(x, y2)
plt.xlabel('n ---> ')
plt.ylabel('(0.2)[1.2]^n')
# plt.show()

plt.subplot(413)
plt.stem(x, y3)
plt.xlabel('n ---> ')
plt.ylabel('(-0.8)^(n^4)')
# plt.show()

plt.subplot(414)
plt.stem(x, y4)
plt.xlabel('n ---> ')
plt.ylabel('-4(0.8)^n')

plt.subplots_adjust(hspace=2)
plt.show()
```

Question 3

```python
n = range(-10, 10)

yn = [np.exp((-1/12 + 1j * (np.pi/6)) * y) for y in n]

real = [element.real for element in yn]
imag = [element.imag for element in yn]

plt.subplot(211)
plt.stem(real, n)
plt.title('Real Part')
plt.xlabel('n')
plt.ylabel('real')

plt.subplot(212)
plt.stem(imag, n)
plt.title('Imaginary Part')
plt.xlabel('n')
plt.ylabel('imag')

plt.subplots_adjust(hspace=2)
plt.show()
```

Part a

```python
x_val = range(-10, 10)
n = x_val

a = np.cos([0.2*math.pi*i for i in n]) 
b = np.cos([1.8*math.pi*i for i in n]) 
c = np.cos([2.2*math.pi*i for i in n]) 

functions = [a, b, c]

for y in functions:
    plt.plot(x_val, y, '.r-')
    plt.axis([-10, 10, -1.1, 1.1])
    plt.xlabel('n ---> ')

    plt.ylabel('Amplitude --->')
    plt.show()
```

```python
x_val = range(-20, 20)
n = x_val

a = np.cos([4 * math.pi * i/17 for i in n]) 
b = 3*np.cos([1.3*math.pi*i for i in n]) - 4*np.sin([0.5*math.pi*i + 0.5*math.pi for i in n])
c = 5*np.cos([1.5*math.pi*i + 0.75*math.pi for i in n]) - 4*np.cos([0.6*math.pi*i for i in n]) - np.sin([0.5*math.pi*i for i in n])

functions = [a, b, c]

for y in functions:
    plt.plot(x_val, y, '.r-')
    plt.axis([-20, 20, -7, 7])
    plt.xlabel('n ---> ')

    plt.ylabel('Amplitude --->')
    plt.show()
```

```python
def unit_step(n, a=0):
    return [1 if x >= a else 0 for x in n]
```

Question 7

```python
N = np.arange(-4, 15)

a = unit_step(N, 5)
b = unit_step(N, 10)
x_n = [0.5*a[i] - 0.5*b[i] for i in range(len(a))]

a = unit_step(N, 0)
b = unit_step(N, 2)
print(a)
print(N)
print(b)
h_n = [a[i] - b[i] for i in range(len(a))]
y_n = scipy.signal.convolve(x_n, h_n, mode='same')

plt.stem(N,x_n)
plt.title('0.5(u[n-5] - u[n-10])')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.xticks(N)
plt.show()

plt.stem(N, h_n)
plt.title('u[n] - u[n-2]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.xticks(N)
plt.show()

plt.stem(N, y_n)
plt.title('x[n] * h[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.xticks(N)
plt.show()
```
